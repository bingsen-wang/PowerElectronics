#Positive Sequence and Negative Sequence of Three-Phase Variables

In three-phase power systems, the voltages are typically sinusoidal. The three phase voltages $v_{ap},v_{bp},v_{cp}$ form a positive sequence if they have the same amplitude and frequency and different phase angles with $v_{ap}$ leading $v_{bp}$ by 120 degrees and $v_{bp}$ leading $v_{cp}$ by 120 degrees.
$$\begin{align}
v_{ap} &= V_p\cos\left(\omega_e t+\theta_p \right)\\
v_{bp} &= V_p\cos\left(\omega_e t+\theta_p - \frac{2\pi}{3} \right)\\
v_{cp} &= V_p\cos\left(\omega_e t+\theta_p + \frac{2\pi}{3} \right)
\end{align}\tag{1}$$

The three phase voltages $v_{an},v_{bn},v_{cn}$ form a negative sequence if they have the same amplitude and frequency and different phase angles with $v_{an}$ lagging $v_{bn}$ by 120 degrees and $v_{bn}$ lagging $v_{cn}$ by 120 degrees.
$$\begin{align}
v_{an} &= V_n\cos\left(\omega_e t+\theta_n \right)\\
v_{bn} &= V_n\cos\left(\omega_e t+\theta_n + \frac{2\pi}{3} \right)\\
v_{cn} &= V_n\cos\left(\omega_e t+\theta_n - \frac{2\pi}{3} \right)
\end{align}\tag{2}$$

#Space Vector

The space vector is complex variable defined as
$$\underline f = {2\over3}\left(f_a + f_be^{j\frac{2\pi}{3}} + f_ce^{j\frac{4\pi}{3}}\right)\tag{3}$$
where $f_a,f_b,f_c$ are the three-phase variables. In electrical systems, the variable $f$ can be voltage $v$, current $i$, or flux linkage $\lambda$. In power electronics systems, the variable $f$ can be the modulation function or switching function.

##Space Vector of Positive Sequence

With reference to (1) and (3), the voltage space vector $\underline v_p$ for the positive sequence is
$$\begin{align}
\underline v_p &= {2\over3}\left(v_{ap} + v_{bp}e^{j\frac{2\pi}{3}} + v_{cp}e^{j\frac{4\pi}{3}}\right)\\
&= {2V_p\over3}\left[\cos\left(\omega_e t+\theta_p\right) + \cos\left(\omega_e t+\theta_p - \frac{2\pi}{3} \right)e^{j\frac{2\pi}{3}} + \cos\left(\omega_e t+\theta_p + \frac{2\pi}{3} \right)e^{j\frac{4\pi}{3}}\right]
\end{align}\tag{4}$$

Based on Euler's identity $e^{jx}=\cos x + j\sin x$, the cosine function can be written in terms exponential functions.
$$\cos x = \frac{e^{jx} + e^{-jx}}{2}\tag{5}$$

Hence, (4) can be rewritten as
$$\begin{align}\underline v_p &= {2V_p\over3\times2}\begin{bmatrix}
e^{j\left(\omega_e t+\theta_p\right)} + e^{j\left(\omega_e t+\theta_p-\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{j\left(\omega_e t+\theta_p+\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}+\\
e^{-j\left(\omega_e t+\theta_p\right)} + e^{-j\left(\omega_e t+\theta_p-\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{-j\left(\omega_e t+\theta_p+\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}
\end{bmatrix}\\
&= V_pe^{j(\omega t+\theta_p)}
\end{align}\tag{6}$$

##Space Vector of Negative Sequence

With reference to (2) and (3), the voltage space vector $\underline v_n$ for the negative sequence is
$$\begin{align}
\underline v_n &= {2\over3}\left(v_{an} + v_{bn}e^{j\frac{2\pi}{3}} + v_{cn}e^{j\frac{4\pi}{3}}\right)\\
&= {2V_n\over3}\left[\cos\left(\omega_e t+\theta_n\right) + \cos\left(\omega_e t+\theta_n + \frac{2\pi}{3} \right)e^{j\frac{2\pi}{3}} + \cos\left(\omega_e t+\theta_n - \frac{2\pi}{3} \right)e^{j\frac{4\pi}{3}}\right]
\end{align}\tag{7}$$

Based on Euler's identity $e^{jx}=\cos x + j\sin x$, the cosine function can be written in terms exponential functions.
$$\cos x = \frac{e^{jx} + e^{-jx}}{2}\tag{8}$$

Hence, (7) can be rewritten as
$$\begin{align}\underline v_n &= {2V_n\over3\times2}\begin{bmatrix}
e^{j\left(\omega_e t+\theta_n\right)} + e^{j\left(\omega_e t+\theta_n+\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{j\left(\omega_e t+\theta_n-\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}+\\
e^{-j\left(\omega_e t+\theta_n\right)} + e^{-j\left(\omega_e t+\theta_n+\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{-j\left(\omega_e t+\theta_n-\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}
\end{bmatrix}\\
&= V_ne^{-j(\omega_e t+\theta_n)}
\end{align}\tag{9}$$

#Rotating Reference Frame

If the reference frame rotates at angular frequency $\omega$, then the space vector in the reference frame is

$$\underline v_{qd} = \underline v_{\alpha\beta}e^{-j\omega t}\tag{10}$$

Substituting (6) and (9) in (10) yields
$$\underline v_{qd} = V_p e^{j[(\omega_e-\omega) t+\theta_p]} +V_n e^{-j[(\omega_e+\omega) t+\theta_n]}\tag{11}$$

#Python Code
The Python code illustrates the space vector in a rotating reference frame.

```
from operator import eq
import matplotlib
from os import linesep
import numpy as np
from numpy import linspace,cos,sin,pi,exp,angle
import matplotlib.pyplot as plt
import matplotlib.patches as mp
from matplotlib.patches import FancyArrow as fa
from matplotlib import animation,rc
rc('animation', html='jshtml')
plt.rcParams.update({"font.family" : "sans serif","mathtext.fontset" : "cm"})

def sv_pos_alphabeta(V,w,t,theta): #space vector of positive sequence
  return V*exp(1j*(w*t+theta))
def sv_neg_alphabeta(V,w,t,theta): #space vector of negative sequence
  return V*exp(-1j*(w*t+theta))
def ab2dq(sv,w,t):
  return sv*exp(-1j*w*t)
#parameters
Nfps = 600
fe= 1 #frequency in Hz
we=2*pi*fe #rotating frequency
t=linspace(0,1/fe,Nfps)
V = 3 #voltage in V
Vp = .75*V #positive sequence voltage
Vn = V-Vp #negative sequence voltage
theta_p = 2*pi/12 #positive sequence phase
theta_n = -6*pi/12 #negative sequence phase
w = 1*2*pi #reference frame rotational speed
wt = w*t
vp_ab_ser=sv_pos_alphabeta(Vp,we,t,theta_p)
vn_ab_ser=sv_neg_alphabeta(Vn,we,t,theta_n)
v_ab_ser=vp_ab_ser+vn_ab_ser
vp_qd_ser=ab2dq(vp_ab_ser,w,t)
vn_qd_ser=ab2dq(vn_ab_ser,w,t)
v_qd_ser=vp_qd_ser+vn_qd_ser
xymax = 4.5

xylim=[-xymax*1.1,xymax*1.1]
clst=['tab:red','tab:blue','tab:brown','tab:green','tab:purple','tab:olive','tab:orange','k','w'] #colors light bg
title_lst = ['Stationary Ref Frame','Rotating Ref Frame']
x_axis_names = [r'$\alpha$','$d$']
y_axis_names = [r'$\beta$','$q$']
eq_lst=[r'$v\underbar_{\alpha\beta} = Ve^{j(\omega_et+\theta)}$',r'$v\underbar_{qd} = v\underbar_{\alpha\beta} e^{-j\omega t}$']
fig = plt.figure(figsize=(9,16))
fig.tight_layout()
ax_lst = []
for k in range(2):
  ax_lst.append(fig.add_axes((0.5/9,.5,8/9,.5),xlim=xylim,ylim=xylim,fc='none'))
  ax_lst[-1].axis('off')
  ax_lst.append(fig.add_axes((0.5/9,0,8/9,.5),xlim=xylim,ylim=xylim,fc='none'))
  ax_lst[-1].axis('off')
for k in range(2):
  ax_lst[k].annotate("",(xymax,0),(-xymax,0),arrowprops={'arrowstyle':"->",'lw':2})
  ax_lst[k].annotate("",(0,xymax),(0,-xymax),arrowprops={'arrowstyle':"->",'lw':2})
  ax_lst[k].text(xymax,-.03,x_axis_names[k],size=26,va='top',ha='right') #x-axis label
  ax_lst[k].text(0.01,xymax,y_axis_names[k],size=26,va='top') #y label
  ax_lst[k].text(-xymax*.5,xymax*.9,title_lst[k],size=22,ha='center')
  ax_lst[k].text(xymax*.5,xymax*.9,eq_lst[k],size=28,ha='center')
text_omega = ax_lst[0].text(-xymax*.75,xymax*.5,r'', size=30,color='k',ha='center')
lines = []
for k in range(2):
  lines.append(ax_lst[0].plot([],[],'--',lw=1,color=clst[k])[0])
  lines.append(ax_lst[1].plot([],[],'--',lw=1,color=clst[k])[0])
lines.append(ax_lst[0].plot([],[],'-',lw=3,color=clst[3])[0]) #v_ab traj
lines.append(ax_lst[0].plot([],[],'--',lw=1,color=clst[4])[0]) #vp_ab traj
lines.append(ax_lst[0].plot([],[],'--',lw=1,color=clst[5])[0]) #vn_ab traj
lines.append(ax_lst[0].plot([],[],'--',lw=1,color=clst[-2])[0]) #vector sum
lines.append(ax_lst[1].plot([],[],'-',lw=3,color=clst[3])[0]) #v_qd traj
lines.append(ax_lst[1].plot([],[],'--',lw=1,color=clst[4])[0]) #vp_qd traj
lines.append(ax_lst[1].plot([],[],'--',lw=1,color=clst[5])[0]) #vn_qd traj
lines.append(ax_lst[1].plot([],[],'--',lw=1,color=clst[-2])[0]) #vector sum

def drawVectors_srf(ax,xymax,v_ab,vp_ab,vn_ab,v_qd,wt,clst): # in stationary reference frame
  theta = angle(v_ab) if angle(v_ab)>0 else angle(v_ab)+2*pi
  theta_p = angle(vp_ab) if angle(vp_ab)>0 else angle(vp_ab)+2*pi
  theta_n = angle(vn_ab) if angle(vn_ab)>0 else angle(vn_ab)+2*pi
  ax.clear()
  ax.set_xlim(xylim)
  ax.set_ylim(xylim)
  ax.axis('off')
  ax.add_patch(fa(0,0,v_ab.real, v_ab.imag, width=.05,length_includes_head=True,
                   color=clst[-3])) #space vector v_ab
  hapos = 'left' if theta<pi/2 or theta>3*pi/2 else 'right'
  ax.text(v_ab.real,v_ab.imag,r'$v\underbar_{\alpha\beta}$',size=35,color=clst[-2],ha=hapos)
  ax.add_patch(fa(0,0,vp_ab.real, vp_ab.imag, width=.04,length_includes_head=True,
                   color=clst[4])) #space vector vp_ab
  hapos = 'left' if theta_p<pi/2 or theta_p>3*pi/2 else 'right'
  ax.text(vp_ab.real,vp_ab.imag,r'$v\underbar_p$', size = 30,color=clst[4],ha=hapos)
  ax.add_patch(fa(0,0,vn_ab.real, vn_ab.imag, width=.04,length_includes_head=True,
                   color=clst[5])) #space vector vn_ab
  hapos = 'left' if theta_n<pi/2 or theta_n>3*pi/2 else 'right'
  ax.text(vn_ab.real,vn_ab.imag,r'$v\underbar_n$', size = 30,color=clst[5],ha=hapos)
  ax.text(v_ab.real,0,r'$v_\alpha$', size = 26,color=clst[-2],ha='center') #v_alpha
  ax.text(0,v_ab.imag,r'$v_\beta$', size = 26,color=clst[-2],va='center') #v_beta
  ax.text(v_qd.real*cos(wt),v_qd.real*sin(wt), r'$v_d$', size = 26,color=clst[1]) #vd
  ax.text(-v_qd.imag*sin(wt),v_qd.imag*cos(wt), r'$v_q$', size = 26,color=clst[1]) #vq
  ax.annotate("",(.9*xymax*cos(wt),.9*xymax*sin(wt)),(-.9*xymax*cos(wt),-.9*xymax*sin(wt)),
              arrowprops={'arrowstyle':"->","lw":1.5,'color':clst[-2],'alpha':0.6}) #d-axis
  ax.annotate("",(-.9*xymax*sin(wt),.9*xymax*cos(wt)),(.9*xymax*sin(wt),-.9*xymax*cos(wt)),
              arrowprops={'arrowstyle':"->","lw":1.5,'color':clst[-2],'alpha':0.6}) #q-axis
  ax.text(.9*xymax*cos(wt),.9*xymax*sin(wt), x_axis_names[1], size = 26,color=clst[-2]) #d-axis name
  ax.text(-.9*xymax*sin(wt),.9*xymax*cos(wt), y_axis_names[1], size = 26,color=clst[-2]) #q-axis name
  return

def drawVectors_rrf(ax,xymax,v_qd,vp_qd,vn_qd,clst): # in rotating reference frame
  ax.clear()
  ax.set_xlim(xylim)
  ax.set_ylim(xylim)
  ax.axis('off')
  theta = angle(v_qd) if angle(v_qd)>0 else angle(v_qd)+2*pi
  theta_p = angle(vp_qd) if angle(vp_qd)>0 else angle(vp_qd)+2*pi
  theta_n = angle(vn_qd) if angle(vn_qd)>0 else angle(vn_qd)+2*pi
  ax.add_patch(fa(0,0,v_qd.real, v_qd.imag, width=.05,length_includes_head=True,
                   color=clst[-3])) #space vector v_qd
  hapos = 'left' if theta<pi/2 or theta>3*pi/2 else 'right'
  ax.text(v_qd.real,v_qd.imag,r'$v\underbar_{qd}$', size = 35,color=clst[-2],ha=hapos)
  ax.add_patch(fa(0,0,vp_qd.real, vp_qd.imag, width=.04,length_includes_head=True,
                   color=clst[4])) #space vector vp_qd
  hapos = 'left' if theta_p<pi/2 or theta_p>3*pi/2 else 'right'
  ax.text(vp_qd.real,vp_qd.imag,r'$v\underbar_p$', size = 30,color=clst[4],ha=hapos)
  ax.add_patch(fa(0,0,vn_qd.real, vn_qd.imag, width=.04,length_includes_head=True,
                   color=clst[5])) #space vector vn_qd
  hapos = 'left' if theta_n<pi/2 or theta_n>3*pi/2 else 'right'
  ax.text(vn_qd.real,vn_qd.imag,r'$v\underbar_n$', size = 30,color=clst[5],ha=hapos)
  ax.text(v_qd.real,0,r'$v_d$', size = 26,color=clst[1],ha='center') #v_d
  ax.text(0,v_qd.imag,r'$v_q$', size = 26,color=clst[1],va='center') #v_q
  return

# animation function. This is called sequentially
def animate(i):
  # i=300
  if i%30==0: print(i) #for monitoring the progress
  lines[0].set_data([0,v_ab_ser[i].real,v_ab_ser[i].real],
   [v_ab_ser[i].imag,v_ab_ser[i].imag,0]) # dotted line for v_alpha and v_beta
  lines[2].set_data([v_qd_ser[i].real*cos(wt[i]),v_ab_ser[i].real,-v_qd_ser[i].imag*sin(wt[i])],
   [v_qd_ser[i].real*sin(wt[i]),v_ab_ser[i].imag,v_qd_ser[i].imag*cos(wt[i])]) # dotted line for v_qd
  lines[3].set_data([0,v_qd_ser[i].real,v_qd_ser[i].real],
   [v_qd_ser[i].imag,v_qd_ser[i].imag,0]) # dotted line for v_d and v_q
  lines[4].set_data(v_ab_ser[:i+1].real,v_ab_ser[:i+1].imag) # v_ab traj
  lines[5].set_data(vp_ab_ser[:i+1].real,vp_ab_ser[:i+1].imag) # vp_ab traj
  lines[6].set_data(vn_ab_ser[:i+1].real,vn_ab_ser[:i+1].imag) # vn_ab traj
  lines[7].set_data([vn_ab_ser[i].real,v_ab_ser[i].real,vp_ab_ser[i].real],
   [vn_ab_ser[i].imag,v_ab_ser[i].imag,vp_ab_ser[i].imag]) # vector sum
  lines[8].set_data(v_qd_ser[:i+1].real,v_qd_ser[:i+1].imag) # v_qd traj
  lines[10].set_data(vn_qd_ser[:i+1].real,vn_qd_ser[:i+1].imag) # vn_ab traj
  lines[11].set_data([vn_qd_ser[i].real,v_qd_ser[i].real,vp_qd_ser[i].real],
   [vn_qd_ser[i].imag,v_qd_ser[i].imag,vp_qd_ser[i].imag]) # vector sum
  drawVectors_srf(ax_lst[2],xymax,v_ab_ser[i],vp_ab_ser[i],vn_ab_ser[i],v_qd_ser[i],wt[i],clst)
  drawVectors_rrf(ax_lst[3],xymax,v_qd_ser[i],vp_qd_ser[i],vn_qd_ser[i],clst)
  return
anim = animation.FuncAnimation(fig, animate, frames=Nfps, interval=50)
fn = r"SpaceVector_RotatingReferenceFrameUnbalanced.mp4"
writervideo = animation.FFMpegWriter(fps=30)
anim.save(fn, writer=writervideo,dpi = 200)
```
