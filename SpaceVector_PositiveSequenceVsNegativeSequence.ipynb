#Positive Sequence and Negative Sequence of Three-Phase Variables

In three-phase power systems, the voltages are typically sinusoidal. The three phase voltages $v_{ap},v_{bp},v_{cp}$ form a positive sequence if they have the same amplitude and frequency and different phase angles with $v_{ap}$ leading $v_{bp}$ by 120 degrees and $v_{bp}$ leading $v_{cp}$ by 120 degrees.
$$\begin{align}
v_{ap} &= V_p\cos\left(\omega t+\theta_p \right)\\
v_{bp} &= V_p\cos\left(\omega t+\theta_p - \frac{2\pi}{3} \right)\\
v_{cp} &= V_p\cos\left(\omega t+\theta_p + \frac{2\pi}{3} \right)
\end{align}\tag{1}$$

The three phase voltages $v_{an},v_{bn},v_{cn}$ form a negative sequence if they have the same amplitude and frequency and different phase angles with $v_{an}$ lagging $v_{bn}$ by 120 degrees and $v_{bn}$ lagging $v_{cn}$ by 120 degrees.
$$\begin{align}
v_{an} &= V_n\cos\left(\omega t+\theta_n \right)\\
v_{bn} &= V_n\cos\left(\omega t+\theta_n + \frac{2\pi}{3} \right)\\
v_{cn} &= V_n\cos\left(\omega t+\theta_n - \frac{2\pi}{3} \right)
\end{align}\tag{2}$$

#Space Vector

The space vector is complex variable defined as
$$\underline f = {2\over3}\left(f_a + f_be^{j\frac{2\pi}{3}} + f_ce^{j\frac{4\pi}{3}}\right)\tag{3}$$
where $f_a,f_b,f_c$ are the three-phase variables. In electrical systems, the variable $f$ can be voltage $v$, current $i$, or flux linkage $\lambda$. In power electronics systems, the variable $f$ can be the modulation function or switching function.

##Space Vector of Positive Sequence

With reference to (1) and (3), the voltage space vector $\underline v_p$ for the positive sequence is
$$\begin{align}
\underline v_p &= {2\over3}\left(v_{ap} + v_{bp}e^{j\frac{2\pi}{3}} + v_{cp}e^{j\frac{4\pi}{3}}\right)\\
&= {2V_p\over3}\left[\cos\left(\omega t+\theta_p\right) + \cos\left(\omega t+\theta_p - \frac{2\pi}{3} \right)e^{j\frac{2\pi}{3}} + \cos\left(\omega t+\theta_p + \frac{2\pi}{3} \right)e^{j\frac{4\pi}{3}}\right]
\end{align}\tag{4}$$

Based on Euler's identity $e^{jx}=\cos x + j\sin x$, the cosine function can be written in terms exponential functions.
$$\cos x = \frac{e^{jx} + e^{-jx}}{2}\tag{5}$$

Hence, (4) can be rewritten as
$$\begin{align}\underline v_p &= {2V_p\over3\times2}\begin{bmatrix}
e^{j\left(\omega t+\theta_p\right)} + e^{j\left(\omega t+\theta_p-\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{j\left(\omega t+\theta_p+\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}+\\
e^{-j\left(\omega t+\theta_p\right)} + e^{-j\left(\omega t+\theta_p-\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{-j\left(\omega t+\theta_p+\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}
\end{bmatrix}\\
&= V_pe^{j(\omega t+\theta_p)}
\end{align}\tag{6}$$

##Space Vector of Negative Sequence

With reference to (2) and (3), the voltage space vector $\underline v_n$ for the negative sequence is
$$\begin{align}
\underline v_n &= {2\over3}\left(v_{an} + v_{bn}e^{j\frac{2\pi}{3}} + v_{cn}e^{j\frac{4\pi}{3}}\right)\\
&= {2V_n\over3}\left[\cos\left(\omega t+\theta_n\right) + \cos\left(\omega t+\theta_n + \frac{2\pi}{3} \right)e^{j\frac{2\pi}{3}} + \cos\left(\omega t+\theta_n - \frac{2\pi}{3} \right)e^{j\frac{4\pi}{3}}\right]
\end{align}\tag{7}$$

Based on Euler's identity $e^{jx}=\cos x + j\sin x$, the cosine function can be written in terms exponential functions.
$$\cos x = \frac{e^{jx} + e^{-jx}}{2}\tag{8}$$

Hence, (7) can be rewritten as
$$\begin{align}\underline v_n &= {2V_n\over3\times2}\begin{bmatrix}
e^{j\left(\omega t+\theta_n\right)} + e^{j\left(\omega t+\theta_n+\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{j\left(\omega t+\theta_n-\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}+\\
e^{-j\left(\omega t+\theta_n\right)} + e^{-j\left(\omega t+\theta_n+\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{-j\left(\omega t+\theta_n-\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}
\end{bmatrix}\\
&= V_ne^{-j(\omega t+\theta_n)}
\end{align}\tag{9}$$

#Phasors

Phasors take the amplitude and phase of the corresponding sinusoidal waveform in cosine function and form a complex variable. In power systems, the phasors use root-mean-square (RMS) value as the magnitude.

##Phasors for the Positive Sequence and Negative Sequence

Using amplitudes, the phasors for the positive-sequence voltages in (1) are
$$\begin{align}
\hat V_{ap} &= V_pe^{j\theta_p}\\
\hat V_{bp} &= V_pe^{j\left(\theta_p-\frac{2\pi}{3}\right)}\\
\hat V_{cp} &= V_pe^{j\left(\theta_p+\frac{2\pi}{3}\right)}
\end{align}\tag{10}$$

the phasors for the negative-sequence voltages in (2) are
$$\begin{align}
\hat V_{an} &= V_ne^{j\theta_n}\\
\hat V_{bn} &= V_ne^{j\left(\theta_n+\frac{2\pi}{3}\right)}\\
\hat V_{cn} &= V_ne^{j\left(\theta_n-\frac{2\pi}{3}\right)}
\end{align}\tag{11}$$

#Python Code
The Python code illustrates the Clarke transformation and inverse.

```
from operator import eq
import matplotlib
from os import linesep
import numpy as np
from numpy import linspace,cos,sin,pi,exp,angle,array
import matplotlib.pyplot as plt
import matplotlib.patches as mp
from matplotlib.patches import FancyArrow as fa
from matplotlib import animation,rc,transforms
rc('animation', html='jshtml')
plt.rcParams.update({"font.family" : "sans serif","mathtext.fontset" : "cm"})

def v_abc(V,w,t,theta,seq): #voltages in abc
  if seq=='P':
    return V*array([cos(w*t+theta-k*2*pi/3) for k in range(3)])
  elif seq=='N':
    return V*array([cos(w*t+theta+k*2*pi/3) for k in range(3)])
def vabc_phasors(V,theta,seq): #phasors in abc
  if seq=='P':
    return V*array([exp(1j*(theta-k*2*pi/3)) for k in range(3)])
  elif seq=='N':
    return V*array([exp(1j*(theta+k*2*pi/3)) for k in range(3)])
def abc2ab(abc): #Clarke transformation, abc to alpha-beta
  x=0
  for k in range(3): x+=2/3*abc[k]*exp(1j*pi*k*2/3)
  return x

#parameters
Nfps = 600
V,fe,theta_p,theta_n = 3,1,pi/6,pi/4 #amplitude,frequency,phase pos, phase neg
t=linspace(0,1/fe,Nfps)
wt = 2*pi*fe*t #angular frequency
vp_abc=v_abc(V,2*pi*fe,t,theta_p,'P') #vabc positive sequence
vp_phasors=vabc_phasors(V,theta_p,'P') #phasors positive sequence
vp_ab=abc2ab(vp_abc) #v_alphabeta positive sequency
vn_abc=v_abc(V,2*pi*fe,t,theta_n,'N') #vabc negative sequence
vn_phasors=vabc_phasors(V,theta_n,'N') #phasors negative sequence
vn_ab=abc2ab(vn_abc) #v_alphabeta negative sequency

#plotting
xymax,xylim = 4,[-4.5,4.5]
clst=['tab:red','tab:blue','tab:brown','tab:purple','tab:olive','tab:green','k','w'] #colors light bg
tns = [r'$\alpha$',r'$\beta$', '$a$','$b$','$c$',
       r'$v_{ap}$',r'$v_{bp}$',r'$v_{cp}$',
       r'$v_{an}$',r'$v_{bn}$',r'$v_{cn}$',
       r'$\hat V_{ap}$',r'$\hat V_{bp}$',r'$\hat V_{cp}$',
       r'$\hat V_{an}$',r'$\hat V_{bn}$',r'$\hat V_{cn}$',
       r'$v\underbar_p$',r'$v\underbar_n$',
       ] #labels
fig = plt.figure(figsize=(9,9))
axs,lines,txts = [],[],[]
as1 = {'head_width':.06,'length_includes_head':True,'overhang':.5} #arr style
as2 = {'width':.06,'length_includes_head':True} #arr style
as3 = {'width':.03,'length_includes_head':True,'overhang':.4} #arr style
as4 = {'width':.02,'head_width':.08,'length_includes_head':True, 'overhang':-.4} #arr style for iabc vectors
as5 = {'width':.03,'length_includes_head':True,'overhang':-.6} #arr style
for k in range(2):
  axs.append(fig.add_axes((0,0,1,1),xlim=xylim,ylim=xylim,fc='none'))
  rot=transforms.Affine2D().rotate_around(0,0,pi/2*k)+axs[0].transData
  axs[0].add_patch(fa(0,0,xymax,0,**as1,transform=rot,color=clst[-2],alpha=.6))#ab ax
  axs[0].text((1-k)*xymax,k*xymax,tns[k],size=26) #ax labels
axs.append(fig.add_axes((0,0,1,1),xlim=xylim,ylim=xylim,fc='none')) #draw phasors
axs[0].set_axis_off()
axs[2].set_axis_off()
for k in range(3):
  rot=transforms.Affine2D().rotate_around(0,0,2*pi/3*k-pi/2)+axs[0].transData
  lines.append(axs[0].plot([],[],'-',lw=2,transform=rot,color=clst[k],alpha=0.4)[0])
  lines.append(axs[0].plot([],[],'--',lw=1.5,transform=rot,color=clst[k],alpha=0.8)[0])
  lines.append(axs[0].plot([],[],'--',lw=1.5,color=clst[k],alpha=0.8)[0])
  lines.append(axs[0].plot([],[],'-',lw=1.5,color=clst[k])[0])
  txts.append(axs[0].text(0,0,tns[5+k],size=30,transform=rot,color=clst[k])) #vabc text
  rot=transforms.Affine2D().rotate_around(0,0,k*2*pi/3)+axs[0].transData
  axs[0].add_patch(fa(0,0,.9*xymax,0,**as1,transform=rot,color=clst[k],alpha=.6))#abc ax
  axs[0].text(.9*xymax,0,tns[k+2],size=25,transform=rot,color=clst[k]) #ax labels
  axs[2].add_patch(fa(0,0,vp_phasors[k].real,vp_phasors[k].imag,**as4,color=clst[k])) #phasors
  axs[2].text(vp_phasors[k].real,vp_phasors[k].imag,tns[k+11],size=25,color=clst[k]) #phasor labels
lines.append(axs[0].plot([],[],'-',lw=3,color=clst[3])[0]) #v_alpha, v_beta

def drawVectors_srf(ax,xymax,vabc,vab,wt,seq,clst): # in stationary reference frame
  ax.clear()
  ax.set_xlim(xylim)
  ax.set_ylim(xylim)
  ax.axis('off')
  ax.add_patch(fa(0,0,vab.real*1.5, vab.imag*1.5, **as2,color=clst[-3])) #scaled
  ax.add_patch(fa(0,0,vab.real, vab.imag, **as3,color=clst[-2])) #sv_ab
  if seq == 0:
    ax.text(vab.real*1.5,vab.imag*1.5,r'$1.5v\underbar_p$',size = 30,color=clst[-3])
    ax.text(vab.real,vab.imag,r'$v\underbar_p$',size = 35,color=clst[-2])
  else:
    ax.text(vab.real*1.5,vab.imag*1.5,r'$1.5v\underbar_n$',size = 30,color=clst[-3])
    ax.text(vab.real,vab.imag,r'$v\underbar_n$',size = 35,color=clst[-2])

  for k in range(3):
    rot=transforms.Affine2D().rotate_around(0,0,2*pi/3*k)+ax.transData
    ax.add_patch(fa(0,0,vabc[k],0,**as3,transform=rot,color=clst[k]))
    ax.add_patch(mp.Rectangle((vabc[k],0),0.1,0.1,angle=-90*np.heaviside(vabc[k],0)-90,
                              rotation_point='xy',transform=rot,color=clst[k],fc='none'))
  return

# animation function. This is called sequentially
def animate(i):
  # i=199
  if i%30==0: print(i) #for monitoring the progress
  ii = int(i/Nfps)
  i = i%Nfps
  if ii==0:
    vabc = vp_abc
    vab = vp_ab
  else:
    if i==0: #clear axis, redraw phasors, change the text
      axs[2].clear()
      axs[2].set_xlim(xylim)
      axs[2].set_ylim(xylim)
      axs[2].axis('off')
      for k in range(3):
        axs[2].add_patch(fa(0,0,vn_phasors[k].real,vn_phasors[k].imag,**as4,color=clst[k])) #phasors
        axs[2].text(vn_phasors[k].real,vn_phasors[k].imag,tns[k+14],size=25,color=clst[k]) #phasor labels
        txts[k].set_text(tns[8+k])
    vabc = vn_abc
    vab = vn_ab
  for k in range(3):
    lines[4*k].set_data(.6*t[:i+1]/t[-1]*xymax,vabc[k,:i+1]) #va,b,c
    lines[4*k+1].set_data([0,.6*t[i]/t[-1]*xymax],[vabc[k,i],vabc[k,i]])
    lines[4*k+2].set_data([vabc[k,i]*cos(k*2*pi/3),vab[i].real],[vabc[k,i]*sin(k*2*pi/3),vab[i].imag])
    txts[k].set_position((.6*t[i]/t[-1]*xymax,vabc[k,i]*2/3))
    if k>0:
      v2 = 0
      for kk in range(k+1):
        v2+=vabc[kk,i]*exp(1j*pi*kk*2/3)
      v1=v2 - vabc[k,i]*exp(1j*pi*k*2/3)
      lines[4*k+3].set_data([v1.real,v2.real],[v1.imag,v2.imag])
  lines[12].set_data(vab[:i+1].real,vab[:i+1].imag)
  drawVectors_srf(axs[1],xymax,vabc[:,i],vab[i],wt[i],ii,clst)
  return
anim = animation.FuncAnimation(fig, animate, frames=Nfps*2, interval=50)
fn = r"SpaceVector_PositiveVsNegativeSequence.mp4"
writervideo = animation.FFMpegWriter(fps=50)
anim.save(fn, writer=writervideo,dpi = 200)
```
