#Positive-Sequence Fundamental and Negative-Sequence 5th-Order Harmonic

In three-phase power systems, the voltages are typically sinusoidal. The three phase voltages $v_{a1},v_{b1},v_{c1}$ form a positive sequence if they have the same amplitude and frequency and different phase angles with $v_{a1}$ leading $v_{b1}$ by 120 degrees and $v_{b1}$ leading $v_{c1}$ by 120 degrees.
$$\begin{align}
v_{a1} &= V_1\cos\left(\omega_e t+\theta_1 \right)\\
v_{b1} &= V_1\cos\left(\omega_e t+\theta_1 - \frac{2\pi}{3} \right)\\
v_{c1} &= V_1\cos\left(\omega_e t+\theta_1 + \frac{2\pi}{3} \right)
\end{align}\tag{1}$$

The three-phase 5th-order voltages $v_{a5},v_{b5},v_{c5}$ form a negative sequence if they have the same amplitude and frequency and different phase angles with $v_{a5}$ lagging $v_{b5}$ by 120 degrees and $v_{b5}$ lagging $v_{c5}$ by 120 degrees.
$$\begin{align}
v_{a5} &= V_5\cos\left(5\omega_e t+\theta_5 \right)\\
v_{b5} &= V_5\cos\left(5\omega_e t+\theta_5 + \frac{2\pi}{3} \right)\\
v_{c5} &= V_5\cos\left(5\omega_e t+\theta_5 - \frac{2\pi}{3} \right)
\end{align}\tag{2}$$

The three-phase 7th-order voltages $v_{a7},v_{b7},v_{c7}$ form a negative sequence if they have the same amplitude and frequency and different phase angles with $v_{a7}$ leading $v_{b7}$ by 120 degrees and $v_{b7}$ leading $v_{c7}$ by 120 degrees.
$$\begin{align}
v_{a7} &= V_7\cos\left(7\omega_e t+\theta_7 \right)\\
v_{b7} &= V_7\cos\left(7\omega_e t+\theta_7 - \frac{2\pi}{3} \right)\\
v_{c7} &= V_7\cos\left(7\omega_e t+\theta_7 + \frac{2\pi}{3} \right)
\end{align}\tag{3}$$

#Space Vector

The space vector is complex variable defined as
$$\underline f = {2\over3}\left(f_a + f_be^{j\frac{2\pi}{3}} + f_ce^{j\frac{4\pi}{3}}\right)\tag{4}$$
where $f_a,f_b,f_c$ are the three-phase variables. In electrical systems, the variable $f$ can be voltage $v$, current $i$, or flux linkage $\lambda$. In power electronics systems, the variable $f$ can be the modulation function or switching function.

##Space Vector of Fundamental

With reference to (1) and (4), the voltage space vector $\underline v_1$ for the fundamental positive sequence is
$$\begin{align}
\underline v_1 &= {2\over3}\left(v_{a1} + v_{b1}e^{j\frac{2\pi}{3}} + v_{c1}e^{j\frac{4\pi}{3}}\right)\\
&= {2V_1\over3}\left[\cos\left(\omega_e t+\theta_1\right) + \cos\left(\omega_e t+\theta_1 - \frac{2\pi}{3} \right)e^{j\frac{2\pi}{3}} + \cos\left(\omega_e t+\theta_1 + \frac{2\pi}{3} \right)e^{j\frac{4\pi}{3}}\right]
\end{align}\tag{5}$$

Based on Euler's identity $e^{jx}=\cos x + j\sin x$, the cosine function can be written in terms exponential functions.
$$\cos x = \frac{e^{jx} + e^{-jx}}{2}\tag{6}$$

Hence, (5) can be rewritten as
$$\begin{align}\underline v_1 &= {2V_1\over3\times2}\begin{bmatrix}
e^{j\left(\omega_e t+\theta_1\right)} + e^{j\left(\omega_e t+\theta_1-\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{j\left(\omega_e t+\theta_1+\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}+\\
e^{-j\left(\omega_e t+\theta_1\right)} + e^{-j\left(\omega_e t+\theta_1-\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{-j\left(\omega_e t+\theta_1+\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}
\end{bmatrix}\\
&= V_1e^{j(\omega t+\theta_1)}
\end{align}\tag{7}$$

##Space Vector of $5^{th}$-Order Harmonic

With reference to (2) and (4), the voltage space vector $\underline v_5$ for the negative sequence is
$$\begin{align}
\underline v_5 &= {2\over3}\left(v_{a5} + v_{b5}e^{j\frac{2\pi}{3}} + v_{c5}e^{j\frac{4\pi}{3}}\right)\\
&= {2V_5\over3}\left[\cos\left(5\omega_e t+\theta_5\right) + \cos\left(5\omega_e t+\theta_5 + \frac{2\pi}{3} \right)e^{j\frac{2\pi}{3}} + \cos\left(5\omega_e t+\theta_5 - \frac{2\pi}{3} \right)e^{j\frac{4\pi}{3}}\right]
\end{align}\tag{8}$$

Based on Euler's identity $e^{jx}=\cos x + j\sin x$, the cosine function can be written in terms exponential functions.
$$\cos x = \frac{e^{jx} + e^{-jx}}{2}\tag{9}$$

Hence, (8) can be rewritten as
$$\begin{align}\underline v_5 &= {2V_5\over3\times2}\begin{bmatrix}
e^{j\left(5\omega_e t+\theta_5\right)} + e^{j\left(5\omega_e t+\theta_5+\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{j\left(5\omega_e t+\theta_5-\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}+\\
e^{-j\left(5\omega_e t+\theta_5\right)} + e^{-j\left(5\omega_e t+\theta_5+\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{-j\left(5\omega_e t+\theta_5-\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}
\end{bmatrix}\\
&= V_5e^{-j(5\omega_e t+\theta_5)}
\end{align}\tag{10}$$

##Space Vector of $7^{th}$-Order Harmonic

With reference to (3) and (4), the voltage space vector $\underline v_5$ for the negative sequence is
$$\begin{align}
\underline v_7 &= {2\over3}\left(v_{a7} + v_{b7}e^{j\frac{2\pi}{3}} + v_{c7}e^{j\frac{4\pi}{3}}\right)\\
&= {2V_7\over3}\left[\cos\left(7\omega_e t+\theta_7\right) + \cos\left(7\omega_e t+\theta_7 - \frac{2\pi}{3} \right)e^{j\frac{2\pi}{3}} + \cos\left(7\omega_e t+\theta_7 + \frac{2\pi}{3} \right)e^{j\frac{4\pi}{3}}\right]
\end{align}\tag{11}$$

Based on Euler's identity $e^{jx}=\cos x + j\sin x$, the cosine function can be written in terms exponential functions.
$$\cos x = \frac{e^{jx} + e^{-jx}}{2}\tag{12}$$

Hence, (11) can be rewritten as
$$\begin{align}\underline v_7 &= {2V_7\over3\times2}\begin{bmatrix}
e^{j\left(7\omega_e t+\theta_7\right)} + e^{j\left(7\omega_e t+\theta_7-\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{j\left(7\omega_e t+\theta_7+\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}+\\
e^{-j\left(5\omega_e t+\theta_7\right)} + e^{-j\left(7\omega_e t+\theta_7-\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{-j\left(7\omega_e t+\theta_7+\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}
\end{bmatrix}\\
&= V_7e^{j(7\omega_e t+\theta_7)}
\end{align}\tag{13}$$

#Rotating Reference Frame

If the reference frame rotates at angular frequency $\omega$, then the space vector in the reference frame is

$$\underline v_{qd} = \underline v_{\alpha\beta}e^{-j\omega t}\tag{14}$$

Substituting (7), (10), and (13) in (14) yields
$$\underline v_{qd} = V_1 e^{j[(\omega_e-\omega) t+\theta_1]} +V_5 e^{-j[(5\omega_e+\omega) t+\theta_5]} +  V_7 e^{j[(7\omega_e-\omega) t+\theta_7]}\tag{15}$$

#Python Code
The Python code illustrates the space vector with 5th- and 7th-order harmonics in a synchronous reference frame.

```
from operator import eq
import matplotlib
from os import linesep
import numpy as np
from numpy import linspace,cos,sin,pi,exp,angle
from numpy import heaviside as hvs
import matplotlib.pyplot as plt
import matplotlib.patches as mp
from matplotlib.patches import FancyArrow as fa
from matplotlib import animation,rc
rc('animation', html='jshtml')
plt.rcParams.update({"font.family" : "sans serif","mathtext.fontset" : "cm"})

def sv_pos_alphabeta(V,w,t,theta): #space vector of positive sequence
  return V*exp(1j*(w*t+theta))
def sv_neg_alphabeta(V,w,t,theta): #space vector of negative sequence
  return V*exp(-1j*(w*t+theta))
def ab2dq(sv,w,t):
  return sv*exp(-1j*w*t)
#parameters
Nfps = 900 #number of frames, reduce this number for faster run
fe= 1 #frequency in Hz
we=2*pi*fe #rotating frequency
t=linspace(0,1/fe,Nfps)
V = 3 #voltage in V
V1 = .8*V #fundamental
V5 = .21*V #5th-order
V7 = .15*V #7th-order
theta_1 = 8*pi/12 #phase angle of fundamental
theta_5 = -6*pi/12 #phase angle of 5th-order
theta_7 = 4*pi/12 #phase angle of 7th-order
w = we #reference frame rotational speed
wt = w*t
v1_ab = sv_pos_alphabeta(V1,we,t,theta_1) #fundamental, positive sequence
v5_ab = sv_neg_alphabeta(V5,5*we,t,theta_5)#5th-order, negative sequence
v7_ab = sv_pos_alphabeta(V7,7*we,t,theta_7)#7th-order, postive sequence
v_ab = v1_ab + v5_ab + v7_ab #space vector of fundamental, 5th, 7th
v15_ab = v1_ab + v5_ab #space vector of fundamental, 5th
v1_qd = ab2dq(v1_ab,w,t)
v5_qd = ab2dq(v5_ab,w,t)
v7_qd = ab2dq(v7_ab,w,t)
v_qd = v1_qd + v5_qd + v7_qd #space vector of fundamental, 5th, 7th
v15_qd = v1_qd+v5_qd #space vector of fundamental, 5th
xymax = 4.5

xylim=[-xymax*1.1,xymax*1.1]
clst=['tab:red','tab:blue','tab:brown','tab:green','tab:purple','tab:olive','tab:orange','k','w'] #colors light bg
title_lst = ['Stationary Frame','Synchronous Frame']
x_axis_names = [r'$\alpha$','$d$']
y_axis_names = [r'$\beta$','$q$']
eq_lst=[r'$v\underbar_{\alpha\beta} = V_1e^{j(\omega_et+\theta_1)}$',
        r'$+V_5e^{-j(5\omega_et+\theta_5)}+V_7e^{j(7\omega_et+\theta_7)}$',
        r'$v\underbar_{qd} = v\underbar_{\alpha\beta} e^{-j\omega t}$','']
fig = plt.figure(figsize=(9,16))
fig.tight_layout()
ax_lst = []
for k in range(2):
  ax_lst.append(fig.add_axes((0.5/9,.5,8/9,.5),xlim=xylim,ylim=xylim,fc='none'))
  ax_lst[-1].axis('off')
  ax_lst.append(fig.add_axes((0.5/9,0,8/9,.5),xlim=xylim,ylim=xylim,fc='none'))
  ax_lst[-1].axis('off')
for k in range(2):
  ax_lst[k].annotate("",(xymax,0),(-xymax,0),arrowprops={'arrowstyle':"->",'lw':2})
  ax_lst[k].annotate("",(0,xymax),(0,-xymax),arrowprops={'arrowstyle':"->",'lw':2})
  ax_lst[k].text(xymax,-.03,x_axis_names[k],size=26,va='top',ha='right') #x-axis label
  ax_lst[k].text(0.01,xymax,y_axis_names[k],size=26,va='top') #y label
  ax_lst[k].text(-xymax*.5,xymax*.9,title_lst[k],size=22,ha='center')
  ax_lst[k].text(xymax*.5,xymax*.9,eq_lst[2*k],size=22,ha='center')
  ax_lst[k].text(xymax*.5,xymax*.8,eq_lst[2*k+1],size=18,ha='center')
lines = []
for k in range(2):
  lines.append(ax_lst[0].plot([],[],'--',lw=1,color=clst[k])[0])
  lines.append(ax_lst[1].plot([],[],'--',lw=1,color=clst[k])[0])
lines.append(ax_lst[0].plot([],[],'-',lw=3,color=clst[-3])[0]) #v_ab traj
lines.append(ax_lst[0].plot([],[],'--',lw=1,color=clst[4])[0]) #v1_ab traj
lines.append(ax_lst[0].plot([],[],'--',lw=1,color=clst[5])[0]) #v5_ab traj
lines.append(ax_lst[0].plot([],[],'--',lw=1,color=clst[3])[0]) #v7_ab traj
lines.append(ax_lst[1].plot([],[],'-',lw=3,color=clst[-3])[0]) #v_qd traj
lines.append(ax_lst[1].plot([],[],'--',lw=1,color=clst[4])[0]) #v1_qd traj
lines.append(ax_lst[1].plot([],[],'--',lw=1,color=clst[5])[0]) #v5_qd traj
lines.append(ax_lst[1].plot([],[],'--',lw=1,color=clst[3])[0]) #v7_qd traj

def drawVectors_srf(ax,xymax,v,v1,v5,v7,v_qd,wt,clst): # in stationary reference frame
  ps = ['right','left','top','bottom'] #list of alignment positions
  v15 = v1 + v5 #space vector of fundamental, 5th
  ax.clear()
  ax.set_xlim(xylim)
  ax.set_ylim(xylim)
  ax.axis('off')
  ax.add_patch(fa(0,0,v.real, v.imag, width=.05,length_includes_head=True,
                   color=clst[-2])) #space vector v_ab
  ax.text(v.real,v.imag,r'$v\underbar_{\alpha\beta}$',size=35,color=clst[-2],
          ha=ps[int(hvs(v.real,0))],va=ps[int(hvs(v.imag,0))+2])
  ax.add_patch(fa(0,0,v1.real, v1.imag, width=.04,length_includes_head=True,
                   color=clst[4])) #space vector v1_ab
  ax.text(.5*v1.real,.5*v1.imag,r'$v\underbar_1$', size = 24,color=clst[4])
  ax.add_patch(fa(v1.real,v1.imag,v5.real, v5.imag, width=.04,length_includes_head=True,
                   color=clst[5])) #space vector v5_ab
  ax.text((v1+.5*v5).real,(v1+.5*v5).imag,r'$v\underbar_5$', size = 24,color=clst[5])
  ax.add_patch(fa(v15.real,v15.imag,v7.real, v7.imag, width=.04,length_includes_head=True,
                   color=clst[3])) #space vector v7_ab
  ax.text((v15+.5*v7).real,(v15+.5*v7).imag,r'$v\underbar_7$', size = 24,color=clst[3])
  ax.text(v.real,0,r'$v_\alpha$', size = 26,color=clst[-2],va=ps[int(hvs(-v.imag,0))+2],ha='center') #v_alpha
  ax.text(0,v.imag,r'$v_\beta$', size = 26,color=clst[-2],ha=ps[int(hvs(-v.real,0))],va='center') #v_beta
  ax.text(v_qd.real*cos(wt),v_qd.real*sin(wt), r'$v_d$', size = 26,color=clst[1]) #vd
  ax.text(-v_qd.imag*sin(wt),v_qd.imag*cos(wt), r'$v_q$', size = 26,color=clst[1]) #vq
  ax.annotate("",(.9*xymax*cos(wt),.9*xymax*sin(wt)),(-.9*xymax*cos(wt),-.9*xymax*sin(wt)),
              arrowprops={'arrowstyle':"->","lw":1.5,'color':clst[-2],'alpha':0.6}) #d-axis
  ax.annotate("",(-.9*xymax*sin(wt),.9*xymax*cos(wt)),(.9*xymax*sin(wt),-.9*xymax*cos(wt)),
              arrowprops={'arrowstyle':"->","lw":1.5,'color':clst[-2],'alpha':0.6}) #q-axis
  ax.text(.9*xymax*cos(wt),.9*xymax*sin(wt), x_axis_names[1], size = 26,color=clst[-2]) #d-axis name
  ax.text(-.9*xymax*sin(wt),.9*xymax*cos(wt), y_axis_names[1], size = 26,color=clst[-2]) #q-axis name
  return

def drawVectors_rrf(ax,xymax,v,v1,v5,v7,clst): # in rotating reference frame
  v15 = v1 + v5 #space vector of fundamental, 5th
  ax.clear()
  ax.set_xlim(xylim)
  ax.set_ylim(xylim)
  ax.axis('off')
  ax.add_patch(fa(0,0,v.real, v.imag, width=.05,length_includes_head=True,
                   color=clst[-2])) #space vector v_qd
  ax.text(v.real,v.imag,r'$v\underbar_{qd}$', size = 35,color=clst[-2],ha='right')
  ax.add_patch(fa(0,0,v1.real, v1.imag, width=.04,length_includes_head=True,
                   color=clst[4])) #space vector v1_qd
  ax.text(.5*v1.real,.5*v1.imag,r'$v\underbar_1$', size = 24,color=clst[4],
          ha='right',va='top')
  ax.add_patch(fa(v1.real,v1.imag,v5.real, v5.imag, width=.04,length_includes_head=True,
                   color=clst[5])) #space vector v5_qd
  ax.text((v1+.5*v5).real,(v1+.5*v5).imag,r'$v\underbar_5$',size=24,color=clst[5])
  ax.add_patch(fa(v15.real,v15.imag,v7.real,v7.imag,width=.04,length_includes_head=True,
                   color=clst[3])) #space vector v7_qd
  ax.text((v15+.5*v7).real,(v15+.5*v7).imag,r'$v\underbar_7$',size=24,color=clst[3])
  ax.text(v.real,0,r'$v_d$', size = 26,color=clst[1],ha='center',va='top') #v_d
  ax.text(0,v.imag,r'$v_q$', size = 26,color=clst[1],va='center') #v_q
  return

# animation function. This is called sequentially
def animate(i):
  # i=Nfps-300
  if i%30==0: print(i) #for monitoring the progress
  lines[0].set_data([0,v_ab[i].real,v_ab[i].real],
   [v_ab[i].imag,v_ab[i].imag,0]) # dotted line for v_alpha and v_beta
  lines[2].set_data([v_qd[i].real*cos(wt[i]),v_ab[i].real,-v_qd[i].imag*sin(wt[i])],
   [v_qd[i].real*sin(wt[i]),v_ab[i].imag,v_qd[i].imag*cos(wt[i])]) # dotted line for v_qd
  lines[3].set_data([0,v_qd[i].real,v_qd[i].real],
   [v_qd[i].imag,v_qd[i].imag,0]) # dotted line for v_d and v_q
  lines[4].set_data(v_ab[:i+1].real,v_ab[:i+1].imag) # v_ab traj
  lines[5].set_data(v1_ab[:i+1].real,v1_ab[:i+1].imag) # v1_ab traj
  lines[6].set_data(v5_ab[:i+1].real+v1_ab[i].real,v5_ab[:i+1].imag+v1_ab[i].imag) # v5_ab traj
  lines[7].set_data(v7_ab[:i+1].real+v15_ab[i].real,v7_ab[:i+1].imag+v15_ab[i].imag) # v7_ab traj
  lines[8].set_data(v_qd[:i+1].real,v_qd[:i+1].imag) # v_qd traj
  lines[10].set_data(v5_qd[:i+1].real+v1_qd[i].real,v5_qd[:i+1].imag+v1_qd[i].imag) # v5_qd traj
  lines[11].set_data(v7_qd[:i+1].real+v15_qd[i].real,v7_qd[:i+1].imag+v15_qd[i].imag) # v7_qd traj

  drawVectors_srf(ax_lst[2],xymax,v_ab[i],v1_ab[i],v5_ab[i],v7_ab[i],v_qd[i],wt[i],clst)
  drawVectors_rrf(ax_lst[3],xymax,v_qd[i],v1_qd[i],v5_qd[i],v7_qd[i],clst)
  return
anim = animation.FuncAnimation(fig, animate, frames=Nfps, interval=50)
fn = r"SpaceVector_RotatingReferenceFrame5th7thOrder.mp4"
writervideo = animation.FFMpegWriter(fps=20)
anim.save(fn, writer=writervideo,dpi = 200)
```
