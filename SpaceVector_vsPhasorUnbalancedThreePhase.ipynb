#Positive Sequence and Negative Sequence of Three-Phase Variables

In three-phase power systems, the voltages are typically sinusoidal. The three phase voltages $v_{ap},v_{bp},v_{cp}$ form a positive sequence if they have the same amplitude and frequency and different phase angles with $v_{ap}$ leading $v_{bp}$ by 120 degrees and $v_{bp}$ leading $v_{cp}$ by 120 degrees.
$$\begin{align}
v_{ap} &= V_p\cos\left(\omega t+\theta_p \right)\\
v_{bp} &= V_p\cos\left(\omega t+\theta_p - \frac{2\pi}{3} \right)\\
v_{cp} &= V_p\cos\left(\omega t+\theta_p + \frac{2\pi}{3} \right)
\end{align}\tag{1}$$

The three phase voltages $v_{an},v_{bn},v_{cn}$ form a negative sequence if they have the same amplitude and frequency and different phase angles with $v_{an}$ lagging $v_{bn}$ by 120 degrees and $v_{bn}$ lagging $v_{cn}$ by 120 degrees.
$$\begin{align}
v_{an} &= V_n\cos\left(\omega t+\theta_n \right)\\
v_{bn} &= V_n\cos\left(\omega t+\theta_n + \frac{2\pi}{3} \right)\\
v_{cn} &= V_n\cos\left(\omega t+\theta_n - \frac{2\pi}{3} \right)
\end{align}\tag{2}$$

#Space Vector

The space vector is complex variable defined as
$$\underline f = {2\over3}\left(f_a + f_be^{j\frac{2\pi}{3}} + f_ce^{j\frac{4\pi}{3}}\right)\tag{3}$$
where $f_a,f_b,f_c$ are the three-phase variables. In electrical systems, the variable $f$ can be voltage $v$, current $i$, or flux linkage $\lambda$. In power electronics systems, the variable $f$ can be the modulation function or switching function.

##Space Vector of Positive Sequence

With reference to (1) and (3), the voltage space vector $\underline v_p$ for the positive sequence is
$$\begin{align}
\underline v_p &= {2\over3}\left(v_{ap} + v_{bp}e^{j\frac{2\pi}{3}} + v_{cp}e^{j\frac{4\pi}{3}}\right)\\
&= {2V_p\over3}\left[\cos\left(\omega t+\theta_p\right) + \cos\left(\omega t+\theta_p - \frac{2\pi}{3} \right)e^{j\frac{2\pi}{3}} + \cos\left(\omega t+\theta_p + \frac{2\pi}{3} \right)e^{j\frac{4\pi}{3}}\right]
\end{align}\tag{4}$$

Based on Euler's identity $e^{jx}=\cos x + j\sin x$, the cosine function can be written in terms exponential functions.
$$\cos x = \frac{e^{jx} + e^{-jx}}{2}\tag{5}$$

Hence, (4) can be rewritten as
$$\begin{align}\underline v_p &= {2V_p\over3\times2}\begin{bmatrix}
e^{j\left(\omega t+\theta_p\right)} + e^{j\left(\omega t+\theta_p-\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{j\left(\omega t+\theta_p+\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}+\\
e^{-j\left(\omega t+\theta_p\right)} + e^{-j\left(\omega t+\theta_p-\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{-j\left(\omega t+\theta_p+\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}
\end{bmatrix}\\
&= V_pe^{j(\omega t+\theta_p)}
\end{align}\tag{6}$$

##Space Vector of Negative Sequence

With reference to (2) and (3), the voltage space vector $\underline v_n$ for the negative sequence is
$$\begin{align}
\underline v_n &= {2\over3}\left(v_{an} + v_{bn}e^{j\frac{2\pi}{3}} + v_{cn}e^{j\frac{4\pi}{3}}\right)\\
&= {2V_n\over3}\left[\cos\left(\omega t+\theta_n\right) + \cos\left(\omega t+\theta_n + \frac{2\pi}{3} \right)e^{j\frac{2\pi}{3}} + \cos\left(\omega t+\theta_n - \frac{2\pi}{3} \right)e^{j\frac{4\pi}{3}}\right]
\end{align}\tag{7}$$

Based on Euler's identity $e^{jx}=\cos x + j\sin x$, the cosine function can be written in terms exponential functions.
$$\cos x = \frac{e^{jx} + e^{-jx}}{2}\tag{8}$$

Hence, (7) can be rewritten as
$$\begin{align}\underline v_n &= {2V_n\over3\times2}\begin{bmatrix}
e^{j\left(\omega t+\theta_n\right)} + e^{j\left(\omega t+\theta_n+\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{j\left(\omega t+\theta_n-\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}+\\
e^{-j\left(\omega t+\theta_n\right)} + e^{-j\left(\omega t+\theta_n+\frac{2\pi}{3}\right)}e^{j\frac{2\pi}{3}} + e^{-j\left(\omega t+\theta_n-\frac{2\pi}{3}\right)}e^{j\frac{4\pi}{3}}
\end{bmatrix}\\
&= V_ne^{-j(\omega t+\theta_n)}
\end{align}\tag{9}$$

#Phasors

Phasors take the amplitude and phase of the corresponding sinusoidal waveform in cosine function and form a complex variable. In power systems, the phasors use root-mean-square (RMS) value as the magnitude.

##Phasors for the Positive Sequence and Negative Sequence

Using amplitudes, the phasors for the positive-sequence voltages in (1) are
$$\begin{align}
\hat V_{ap} &= V_pe^{j\theta_p}\\
\hat V_{bp} &= V_pe^{j\left(\theta_p-\frac{2\pi}{3}\right)}\\
\hat V_{cp} &= V_pe^{j\left(\theta_p+\frac{2\pi}{3}\right)}
\end{align}\tag{10}$$

the phasors for the negative-sequence voltages in (2) are
$$\begin{align}
\hat V_{an} &= V_ne^{j\theta_n}\\
\hat V_{bn} &= V_ne^{j\left(\theta_n+\frac{2\pi}{3}\right)}\\
\hat V_{cn} &= V_ne^{j\left(\theta_n-\frac{2\pi}{3}\right)}
\end{align}\tag{11}$$

#Python Code
The Python code illustrates the space vector and phasors for the positive sequence and the negative sequence.

```
import matplotlib
from os import linesep
import numpy as np
from numpy import linspace,cos,sin,pi,exp,angle
import matplotlib.pyplot as plt
import matplotlib.patches as mp
from matplotlib import animation,rc
rc('animation', html='jshtml')
plt.rcParams.update({"font.family" : "sans serif","mathtext.fontset" : "cm"})

def sv(abc):
  return (2/3)*(abc[0] + abc[1]*exp(1j*2*pi/3) + abc[2]*exp(1j*4*pi/3))

def vabc_pos(V,w,t,theta): #positive sequence
  return np.array([V*cos(w*t+theta),V*cos(w*t+theta-2*pi/3),V*cos(w*t+theta+2*pi/3)])

def vabc_neg(V,w,t,theta): #negative sequence
  return np.array([V*cos(w*t+theta),V*cos(w*t+theta+2*pi/3),V*cos(w*t+theta-2*pi/3)])

def phasors_pos(V,theta): #positive sequence
  return np.array([V*exp(1j*theta),V*exp(1j*(theta-2*pi/3)),V*exp(1j*(theta+2*pi/3))])

def phasors_neg(V,theta): #negative sequence
  return np.array([V*exp(1j*theta),V*exp(1j*(theta+2*pi/3)),V*exp(1j*(theta-2*pi/3))])

#parameters
Nfps = 400
f = 1 #frequency in Hz
t=linspace(0,1/f,Nfps)
V = 3 #voltage in V
Vp = .7*V #positive sequence voltage
Vn = V-Vp #negative sequence voltage
theta_p = -pi/12 #positive sequence phase
theta_n = 5*pi/12 #negative sequence phase
theta_lst = [theta_p,theta_n]
vabc_pos_ser = vabc_pos(Vp,2*pi*f,t,theta_p)
vabc_neg_ser = vabc_neg(Vn,2*pi*f,t,theta_n)
vabc_ser = vabc_pos_ser + vabc_neg_ser
phasors_p = phasors_pos(Vp,theta_p)
phasors_n = phasors_neg(Vn,theta_n)
phasors = phasors_p + phasors_n
svp_ser = sv(vabc_pos_ser)
svn_ser = sv(vabc_neg_ser)
sv_ser = sv(vabc_ser)

xymax = 4.5
clst=['tab:red','tab:blue','tab:brown','tab:green','tab:olive','tab:purple','tab:orange','k','w'] #colors light bg
vabc_eq=[r'$v_a=V_p\cos(\omega t+\theta_p) + V_n\cos(\omega t+\theta_n)$',
           r'$v_b=V_p\cos(\omega t+\theta_p-\frac{2\pi}{3}) + V_n\cos(\omega t+\theta_n+\frac{2\pi}{3})$',
           r'$v_c=V_p\cos(\omega t+\theta_p-\frac{4\pi}{3}) + V_n\cos(\omega t+\theta_n+\frac{4\pi}{3})$']
vabc_labels = [r'$v_a$',r'$v_b$',r'$v_c$',r'$v_{ap}$',r'$v_{bp}$',r'$v_{cp}$',r'$v_{an}$',r'$v_{bn}$',r'$v_{cn}$']
phasor_eq=[r'$\hat V_a = V_p e^{j\theta_p}+V_n e^{j\theta_n}$',
           r'$\hat V_b = V_p e^{j(\theta_p-\frac{2\pi}{3})} + V_n e^{j(\theta_n+\frac{2\pi}{3})}$',
           r'$\hat V_c = V_p e^{j(\theta_p+\frac{2\pi}{3})} + V_n e^{j(\theta_n-\frac{2\pi}{3})}$'
           ]
phasor_labels = [r'$\hat V_{ap}$',r'$\hat V_{bp}$',r'$\hat V_{cp}$',
                 r'$\hat V_{an}$',r'$\hat V_{bn}$',r'$\hat V_{cn}$',
                 r'$\hat V_{a}$',r'$\hat V_{b}$',r'$\hat V_{c}$']
fig = plt.figure(figsize=(9,16))
fig.tight_layout()
ax=fig.add_axes((0.05,7/16,.9,.9*9/16),xlim=[-xymax,xymax], ylim=[-xymax,xymax],fc='none') #no fc
ax0=fig.add_axes((0.05,7/16,.9,.9*9/16),xlim=[-xymax,xymax], ylim=[-xymax,xymax],fc='none') #no fc for animated parts
ax1=fig.add_axes((0.05,0,.9,7/16),xlim=[-.1,2*xymax-.1], ylim=[-xymax*7/9,xymax*7/9],fc='none') #no fc
ax.axis('off') #turn off axis frames
ax1.axis('off') #turn off axis frames
ax.annotate("", (.9*xymax, 0), (-.9*xymax, 0),arrowprops={'arrowstyle':"->"}) #x-axis for space vector
ax.annotate("", (0,.9*xymax), (0,-.9*xymax),arrowprops={'arrowstyle':"->"}) #y-axis for space vector
ax1.annotate("", (2*xymax-.2, 0), (-.1, 0),arrowprops={'arrowstyle':"->"}) #x-axis for voltages
ax1.annotate("", (0,xymax*7/9), (0,-xymax*7/9),arrowprops={'arrowstyle':"->"}) #y-axis for voltages
ax.text(.9*xymax,-.03,'$Re$',size=20,va='top',ha='right') #x-axis label
ax.text(0.01,.9*xymax,'$Im$',size=20,va='top') #y label
ax1.text(2*xymax-.1,-.03,r'$\omega t$',size=24,va='top',ha='right') #x-axis label
ax.text(-xymax*.5,xymax*.85,'Space Vector',size=26,ha='center')
ax.text(-xymax*.5,xymax*.7,r'$v\underbar = \frac{2}{3}\left(v_a + v_b e^{j\frac{2\pi}{3}}+ v_c e^{j\frac{4\pi}{3}}\right)$',
        size=26,ha='center',color=clst[-2])
ax.text(xymax*.5,xymax*.85,'Phasors',size=26,ha='center')
text_theta = ax.text(-xymax*.75,xymax*.5,'', size=30,color='k',ha='center')
lines, text_vabc, text_vabc_phasors = [],[],[]
lines.append(ax.plot([], [], '-', lw=3, color=clst[6])[0]) #v_sv trajectory
lines.append(ax.plot([], [], '--', lw=1, color=clst[5])[0]) #vp_sv trajectory
lines.append(ax.plot([], [], '--', lw=1, color=clst[4])[0]) #vn_sv trajectory
lines.append(ax.plot([], [], '--', lw=1, color=clst[-2])[0]) #sv summation
for i in range(3):
  lines.append(ax1.plot([], [], '-', lw=5, color=clst[i])[0]) #v_abc
  lines.append(ax1.plot([], [], '-', lw=2, color=clst[i],alpha=.7)[0]) #v_abc_pos
  lines.append(ax1.plot([], [], '--', lw=2, color=clst[i],alpha=.5)[0]) #v_abc_neg
  ax.text(.5*xymax,xymax*(6.5-i)/9,phasor_eq[i],size=20,ha='center',color=clst[i]) #phasor equations
  ax1.text(0,xymax*(9-i)/9,vabc_eq[i],size=24,va='top',color=clst[i]) #vabc equations
  text_vabc.append(ax1.text(0,0,vabc_labels[i],size=35,va='center',color=clst[i]))
  text_vabc.append(ax1.text(0,0,vabc_labels[i+3],size=24,va='center',color=clst[i]))
  text_vabc.append(ax1.text(0,0,vabc_labels[i+6],size=24,va='center',color=clst[i]))
  ax.add_patch(mp.FancyArrow(0,0,phasors_p[i].real,phasors_p[i].imag,width=.05*Vp/V,
                             length_includes_head=True,color=clst[i],alpha=.7))
  ax.add_patch(mp.FancyArrow(0,0,phasors_n[i].real,phasors_n[i].imag,width=.05*Vn/V,
                             length_includes_head=True,color=clst[i],alpha=.5))
  ax.add_patch(mp.FancyArrow(0,0,phasors[i].real,phasors[i].imag,width=.05,
                             length_includes_head=True,color=clst[i]))
  ax.plot([phasors_n[i].real,phasors[i].real,phasors_p[i].real],
          [phasors_n[i].imag,phasors[i].imag,phasors_p[i].imag],
          '--',lw=1.5,color=clst[i],alpha=.5)

  text_vabc_phasors.append(ax.text(phasors_p[i].real,phasors_p[i].imag,phasor_labels[i],size=20,color=clst[i]))
  text_vabc_phasors.append(ax.text(phasors_n[i].real,phasors_n[i].imag,phasor_labels[i+3],size=20,color=clst[i]))
  text_vabc_phasors.append(ax.text(phasors[i].real,phasors[i].imag,phasor_labels[i+6],size=24,color=clst[i]))
  # text_theta.set_text()
def drawVectors(ax,xymax,v_sv,vp_sv,vn_sv,clst):
  theta = angle(v_sv) if angle(v_sv)>0 else angle(v_sv)+2*pi
  thetap = angle(vp_sv) if angle(vp_sv)>0 else angle(vn_sv)+2*pi
  thetan = angle(vn_sv) if angle(vn_sv)>0 else angle(vn_sv)+2*pi
  ax.clear()
  ax.set_xlim([-xymax,xymax])
  ax.set_ylim([-xymax,xymax])
  ax.axis('off')
  arr = mp.FancyArrow(0,0,v_sv.real, v_sv.imag, width=.05,length_includes_head=True,
                   color=clst[-2])
  ax.add_patch(arr)
  hapos = 'left' if theta<pi/2 or theta>3*pi/2 else 'right'
  ax.text(v_sv.real,v_sv.imag,r'$v\underbar$', size = 40,color=clst[-2],ha=hapos)
  arr = mp.FancyArrow(0,0,vp_sv.real, vp_sv.imag, width=.04,length_includes_head=True,
                   color=clst[5])
  ax.add_patch(arr)
  hapos = 'left' if thetap<pi/2 or thetap>3*pi/2 else 'right'
  ax.text(vp_sv.real,vp_sv.imag,r'$v\underbar_p$', size = 30,color=clst[5],ha=hapos)
  arr = mp.FancyArrow(0,0,vn_sv.real, vn_sv.imag, width=.04,length_includes_head=True,
                   color=clst[4])
  ax.add_patch(arr)
  hapos = 'left' if thetan<pi/2 or thetan>3*pi/2 else 'right'
  ax.text(vn_sv.real,vn_sv.imag,r'$v\underbar_n$', size = 30,color=clst[4],ha=hapos)
  return

# animation function. This is called sequentially
def animate(i):
  # i=Nfps-30
  if i%20==0: print(i) #for monitoring the progress
  lines[0].set_data(sv_ser[:i+1].real,sv_ser[:i+1].imag) #v_sv trajectory
  lines[1].set_data(svp_ser[:i+1].real,svp_ser[:i+1].imag) #vp_sv trajectory
  lines[2].set_data(svn_ser[:i+1].real,svn_ser[:i+1].imag) #vn_sv trajectory
  lines[3].set_data([svp_ser[i].real,sv_ser[i].real,svn_ser[i].real],
                    [svp_ser[i].imag,sv_ser[i].imag,svn_ser[i].imag]) #sv summation
  for k in range(3):
    lines[3*k+4].set_data(t[:i+1]*f*xymax*1.8,vabc_ser[k,:i+1]) #v_abc
    lines[3*k+5].set_data(t[:i+1]*f*xymax*1.8,vabc_pos_ser[k,:i+1]) #v_abc_pos
    lines[3*k+6].set_data(t[:i+1]*f*xymax*1.8,vabc_neg_ser[k,:i+1]) #v_abc_neg
    text_vabc[3*k].set_position((t[i]*f*xymax*1.8,vabc_ser[k,i])) #v_abc lables
    text_vabc[3*k+1].set_position((t[i]*f*xymax*1.8,vabc_pos_ser[k,i])) #v_abc_pos lables
    text_vabc[3*k+2].set_position((t[i]*f*xymax*1.8,vabc_neg_ser[k,i])) #v_abc_neg lables
  drawVectors(ax0,xymax,sv_ser[i],svp_ser[i],svn_ser[i],clst)
  return
anim = animation.FuncAnimation(fig, animate, frames=Nfps, interval=50)
fn = r"SpaceVector_vsPhasorsUnbalanced.mp4"
writervideo = animation.FFMpegWriter(fps=20)
anim.save(fn, writer=writervideo,dpi = 200)
```
